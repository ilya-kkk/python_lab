<h3>Вывести количество записей, у которых в поле Название строка длиннее 30 символов</h3>

```python
import pandas as pd

file_path = "books.csv"  
df = pd.read_csv(file_path, encoding="windows-1251", sep=";")

count = df[df["Название"].str.len() > 30].shape[0]
print(f"Количество записей, у которых длина строки в поле 'Название' больше 30 символов: {count}")
```

<h3>Реализовать поиск книги по автору, использовать ограничение на выдачу в зависимости от варианта.До 2016 года</h3>

```python
def search_books_by_author(author_name, year_limit=2016):
    author_books = df[df["Автор"].str.contains(author_name, case=False, na=False)]
    
    author_books = author_books[ \
        pd.to_datetime(author_books["Дата поступления"], format="%d.%m.%Y %H:%M", errors="coerce") \
        < pd.to_datetime(f"01.01.{year_limit} 00:00", format="%d.%m.%Y %H:%M", errors="coerce")]
    
    if author_books.empty:
        print(f"Нет книг автора '{author_name}' до {year_limit} года.")
    else:
        print(author_books[["Название", "Автор", "Дата поступления"]])


search_books_by_author("Людмила Петрановская")
```

<h3>Реализовать генератор библиографических ссылок вида <автор>. <название> - <год> для 20 записей. Записи выбрать произвольно. Список сохраняется как отдельный файл текстового формата с нумерацией строк.</h3>

```python
df["Дата поступления"] = pd.to_datetime(df["Дата поступления"], format="%d.%m.%Y %H:%M", errors="coerce")

df["Год"] = df["Дата поступления"].dt.year
sample_books = df.sample(n=20, random_state=42)

sample_books_reset = sample_books.reset_index(drop=True)

biblio_list = [
    f"{i+1}. {row['Автор']}. {row['Название']} - {row['Год']}"
    for i, row in sample_books_reset.iterrows()
]
output_file = "bibliography.txt"
with open(output_file, "w", encoding="utf-8") as file:
    file.write("\n".join(biblio_list))

print(f"Библиографические ссылки сохранены в файл {output_file}")
```

<h3>Используя приложенный файл currency.xml, выполнить следующее:
Распарсить файл и извлечь данные, согласно варианту. Выполнить приведения типов по необходимости. Два отдельных списка Name и Value</h3>

```python
import xml.etree.ElementTree as ET

file_path = 'currency.xml'
tree = ET.parse(file_path)
root = tree.getroot()

Name = []
Value = []

for valute in root.findall('Valute'):
    name = valute.find('Name').text 
    value = valute.find('Value').text  

    try:
        value = float(value.replace(',', '.'))  
    except ValueError:
        value = None  

    Name.append(name)
    Value.append(value)


print("Name:", Name)
print("Value:", Value)
```
